```python
# first imports and the function to load files:
import json
import pprint as pp

# you do not need to understand this function. It just loads the file
def load_json_file_named(file_name):
    try: 
        loaded_data = []
        file_location = f"./data/{file_name}"
        with open(file_location, 'r') as file: # or f"data/{file_name}" depending on your files
            loaded_data =  json.load(file)
    except OSError as e:
        print(f"Error. Does the file exist in this folder? {file_location}\n\n {e}")
    return loaded_data
```

```python
# 1. Write a function which will take details of ONE train, which then will return a string 
# which describes one train as a simple detail, like `*** 11:34 - Inverness ------ Platform 2 ***`

def train_as_line(train_info):
    description = f"*** "
    description += f"{train_info['aimed_departure_time']} "
    description += f"{train_info['destination_name']:-<30}"
    description += f"Platform {train_info['platform']:>3}"
    description += f" ***"
    return description

first_train = trains_edinburgh['departures']['all'][1]
print(train_as_line(first_train))
```

```python
train1 = trains_edinburgh['departures']['all'][0]
train2 = trains_edinburgh['departures']['all'][1]
assert train_as_line(train1) == "*** 09:00 Glasgow Queen Street----------Platform  12 ***"
assert train_as_line(train2) == "*** 09:00 London Kings Cross------------Platform   9 ***"
print("yay")
```

```python
# 2. Write another function which takes details of many such trains, 
# and uses above function to create a list of Departures.

def trains_as_lines(many_trains):
    long_string = ""
    for train in many_trains:
        long_string += train_as_line(train)
        long_string += "\n"
    return long_string

result = trains_as_lines(trains_edinburgh['departures']['all'][0:2])
print(result)
result
```

```python
assert trains_as_lines(trains_edinburgh['departures']['all'][0:2]) == \
    "*** 09:00 Glasgow Queen Street----------Platform  12 ***\n*** 09:00 London Kings Cross------------Platform   9 ***\n"

print("worked!")
```

```python
# 3. Adjust above functions so that they are prettier (e.g. use symbols like `***` 
# and tabs `\t` and new lines `\n` and right-left adjusted string
#  formatting to make the layout pretty).

def trains_as_lines_pretty(many_trains, city):
    long_string = ""
    long_string += f"****************{city:*^30}**************\n"
    long_string += f"*                                                          *\n"
    for train in many_trains:
        long_string += f"* {train_as_line(train)} *"
        long_string += "\n"
    long_string += f"*                                                          *\n"
    long_string += f"************************************************************\n"
    return long_string

result = trains_as_lines_pretty(trains_edinburgh['departures']['all'][0:2], trains_edinburgh['station_name'])
print(result)
result
```

Notice we did not write tests for the above, since it grew arms and legs. You could split it into smaller functions that are easier to test, and then work from those:

```python
def header(city_name):
    return f"****************{city_name:*^30}**************\n"

def footer():
    return f"************************************************************\n"

def break_line():
    return f"*                                                          *\n"

# put the text in the middle of 60 characters with a frame of single * around it
def prettify_line(some_line):
    return f"* {some_line: ^56} *\n"
  
assert header("banana")     == '****************************banana**************************\n'
assert footer()             == '************************************************************\n'
assert break_line()         == '*                                                          *\n'
assert prettify_line("yay") == '*                           yay                            *\n'
print("worked") 
              
```

```python
# now we know individual bits are tested we could assemble bigger structure from it
# this way we at can focus on logic (loops, ifs) not the styles.

def trains_as_lines_pretty(many_trains, city):
    long_string = ""
    long_string += header(city)
    long_string += break_line()
    for train in many_trains:
        long_string += prettify_line(train_as_line(train)) 
    long_string += break_line() 
    long_string += footer()
    return long_string

result = trains_as_lines_pretty(trains_edinburgh['departures']['all'][0:2], trains_edinburgh['station_name'])
print(result)
result
```

OOOH!!! some platforms are None (which is python's way to say 'missing'). Notice it is not missing really, instead (annoyingly) it has a value of "None".

And so it cannot be put into stirng formatting, like `f"*** {platform:^10} ***"` because that only works for strings! Let's adjust our function from above, so that it is more robust.

```python
# 1. Write a function which will take details of ONE train, which then will return a string 
# which describes one train as a simple detail, like `*** 11:34 - Inverness ------ Platform 2 ***`

def train_as_line(train_info):
    description = f"*** "
    description += f"{train_info['aimed_departure_time']} "
    description += f"{train_info['destination_name']:-<30}"
     # this is new down
    platform = train_info['platform'] if train_info['platform'] != None else "?"
    description += f"Platform {platform:>3}"
     # this is new up
    description += f" ***"
    return description

train_weird = trains_glasgow_queen_st['departures']['all'][8]
print(train_as_line(train_weird))
```

```python
def small_header(some_text = ""): # if unknown, assume empty string
    return f"{some_text:*^54}\n"

print(small_header())
print(small_header("banana"))
```

```python

def small_line(left = "", right = ""): # if unknown, assume empty string
    return f"* {left:.<25}{right:.>25} *\n"

print(small_line(""))
print(small_line("colour", "yellow"))
```

```python
def detailed_view(one_train):
    description = f""
    
    description += small_header()
    description += small_line(one_train['expected_departure_time'], f"PLATFORM {one_train['platform']}")
    description += small_line(f"{one_train['origin_name']} ->",one_train['destination_name'])
    description += small_line("",one_train['operator_name'])
    description += small_header()

    return description

train1 = trains_edinburgh['departures']['all'][0]
print(detailed_view(train1))
```
