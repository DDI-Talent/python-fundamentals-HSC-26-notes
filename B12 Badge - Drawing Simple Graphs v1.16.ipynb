## Preparing your data for the graph - worked example

Below is a worked example of taking some data and preparing them into a graph. Follow it through, and then see the minitask below. 

````python
# some data
fruit_colors = {"banana":"yellow", 
                "kiwi": "green", 
                "plum": "purple"}

fruits_I_put_in_the_pie = ["banana", "kiwi","banana", "kiwi","banana", "kiwi",
                "kiwi","kiwi","plum","banana", "peach","peach", "quince", "mullberry"]
````

My plan:

1. count fruits (I will use dictionary comprehension. It's like a List Comp, but needs key-value pairs)
2. sort fruits from most to least popular
3. make a pie chart of fruits (amount of a given fruit will be the size of slice)
4. use appropriate fruit colours (with 'poie colour' for fruits of unknow colour)
5. use legend, titles, etc

### 1. get all unique fruits and count them

````python
unique_fruits = list(set(fruits_I_put_in_the_pie))
print(unique_fruits)
````

````python
# note: this is an example of a Dictionary Comprehension. notice {} brackets
# and the fact top line needs key:value 
words = "kiwi","plum","banana", "peach"
word_lengths = {
    word: len(word)
    for word in words
}
print(word_lengths)
````

Do you remember List's .count() function? you use it like this:

`sime_list.count(item_to_count)`

````python
# how many 4s do we have in this list?
[2,3,4,5,6,4].count(4)
````

````python
# ok, now let's use it to count fruits:
fruit_counts = {
    fruit_name: fruits_I_put_in_the_pie.count(fruit_name)
    for fruit_name in unique_fruits
}
print(fruit_counts)
````

### 2. sort the count from most to least popular (this is new, will take some explaining)

````python
# ok, now let's sort fruits. 
````

````python
# how do you sort things in python?
numbers = [3,1,6,4]
print(sorted(numbers))
````

````python
# but if you sort a dictionary, something weird will happen, can you figure out what?
print(sorted(fruit_counts))
````

Yes... It took the ketys and sorted them. This is not perfect because: we want to not lose our values, and we want to sort by value.

To do that we will split the dictionary into pairs of key-value tupples, and then sort by the second part of each tupple (the value). Have a look:

````python
# what about a dictionary? We'd have to figure out which part we're sorting? key? value?
word_lengths = {'kiwi': 4, 'plum': 4, 'banana': 6, 'peach': 5}
print(word_lengths)

# let's split dict into list of tupples, so we can sort by first os second part of the tupple
tupples = list(word_lengths.items())
print(tupples)
# notice tupples are simple two-items lists. first item (key) has index 0, second item (value) has index 1, so...

sorted_tupples_by_name = sorted(tupples, key = lambda tupple: tupple[0])  # by key
print(sorted_tupples_by_name)
sorted_tupples_by_value = sorted(tupples, key = lambda tupple: tupple[1]) # by value (more useful here)
print(sorted_tupples_by_value)

# now we need toi glue it back into a dictionary
word_lengths_sorted = dict(sorted_tupples_by_value)
print(word_lengths_sorted)
````

````python
# in one line it would be this, but hey, it's some seriously ugly spaghetti code!
word_lengths_sorted = dict(sorted(word_lengths.items(), key = lambda tupple: tupple[1]))
print(word_lengths_sorted)
````

````python
# Ok, that code is so complex I want to write it once, and never think about it again
# functions will help: I will extract this into a function sort_dictionary() which takes any dictionary and returns a sorted version. 

def sort_dictionary(some_dict, from_largest = False, by_value = True):
    tupples = list(some_dict.items())
    if (by_value):
        sorted_tupples = sorted(tupples, key = lambda tupple: tupple[1], reverse = from_largest)
    else:
        sorted_tupples = sorted(tupples, key = lambda tupple: tupple[0], reverse = from_largest)
    return dict(sorted_tupples)
    # this could be further improved, but let's move on for now

print(sort_dictionary(word_lengths))
print(sort_dictionary(word_lengths, from_largest = True))
print(sort_dictionary(word_lengths, by_value = False)) # sort alphabetically by name
````

````python
# ok, we're ready. count and sort fruits:

sorted_fruit_counts = sort_dictionary(fruit_counts)
````

### 3. make a pie chart

````python
plt.pie(sorted_fruit_counts.values(), labels = sorted_fruit_counts.keys() )
plt.show()
````

### 4. use appropriate fruit colours (with 'wheat pie colour' for fruits of unknow colour)

````python
fruit_colors_definitions = {"banana":"yellow", 
                "kiwi": "green", 
                "plum": "purple"}

# graph will take fruit names as a list like ["plum", "kiwi", "banana"]
# so we need to also give it colours as a list, like ["purple","green","yellow"]
# for list of colours see https://matplotlib.org/stable/gallery/color/named_colors.html
default_colour = "wheat"

fruit_colors = [
    fruit_colors_definitions.get(fruit, default_colour)
    for fruit in sorted_fruit_counts.keys()
]
print(fruit_colors)
````

````python
plt.pie(sorted_fruit_counts.values(), labels = sorted_fruit_counts.keys(), 
        colors = fruit_colors,
        wedgeprops={"linewidth": 1, "edgecolor": "white"} # this white lie will make reading the graph easier. remove it and see what happens
       ) 
plt.show()
````

### 5. use legend, titles, etc

a) make banana stand out. Explode takes a list of 'nudge distances'. Here we make all distances 0, apart of banana which will be 10% nudged-out
b) add shadow so it is prettier (but harder to read, ha!)

````python
plt.pie(sorted_fruit_counts.values(), 
        labels = sorted_fruit_counts.keys(), 
        colors = fruit_colors,
        wedgeprops={"linewidth": 1, "edgecolor": "white"},
        explode = [ 0.1 if fruit == "banana" else 0 for fruit in sorted_fruit_counts.keys()],
        shadow = True,
        startangle = 90,
        counterclock = False
       ) 
plt.show()
````

## Back to our main fruit quest:

Say we want to show barcharts next to each other, each with a different fruit 'highlighted'.

- we will have 1 column with 3 rows.

````python
# I'll need this function to get a list of colours like ['gray','gray','red','gray']
def colour_for_highlight(items, highlight_this, highlight_colour = 'red', normal_colour = 'gray'):
    return [ 
        highlight_colour if item == highlight_this else normal_colour
        for item in items
    ]

print(colour_for_highlight(["kiwi","plum","banana", "peach"], "plum"))
print(colour_for_highlight(["kiwi","plum","banana", "peach"], "banana", "yellow"))
````

Ok, so now I will create three almost oidentical graphs and put them next to each other. Your task is:

- do you see why they are in this order?
- what happens when you comment out one whole graph?
- can you edit the code below, so they are not one above another, but one next to each other (horisontally)?
- once you're done, can you rewrite this using a function so that all this code is not repeated so many time? (answer is below)

````python
plt.subplot(3,1,1)
plt.bar(sorted_fruit_counts.keys(),
        sorted_fruit_counts.values(),
        color = colour_for_highlight(sorted_fruit_counts.keys(), "banana", "yellow")
       )
plt.title("banana")
plt.ylabel("number of fruit")

plt.subplot(3,1,2)
plt.bar(sorted_fruit_counts.keys(),
        sorted_fruit_counts.values(),
        color = colour_for_highlight(sorted_fruit_counts.keys(), "plum", "purple")
       )
plt.title("plum")
plt.ylabel("number of fruit")

plt.subplot(3,1,3)
plt.bar(sorted_fruit_counts.keys(),
        sorted_fruit_counts.values(),
        color = colour_for_highlight(sorted_fruit_counts.keys(), "quince", "orange")
       )
plt.title("quince")
plt.ylabel("number of fruit")

plt.show()

````

You noticed that the code above is very repeatitive and not DRY.

Do you remember how to replace repetitive code with functions? 

1. find what chunks of code vary by, and what do they have in common.
2. replace everything which varies with a variable. Innitially do it for just one chunk, not the whole thing.
3. write a function which takes those variables

see below

````python
# 1. I found three variables:
which_item = 1
item_to_highlight = "banana"
colour_to_highlight = "yellow"

# 2. here is the code, with those 3 variables substituted:
plt.subplot(3,1,which_item)
plt.bar(sorted_fruit_counts.keys(),
        sorted_fruit_counts.values(),
        color = colour_for_highlight(sorted_fruit_counts.keys(), item_to_highlight, colour_to_highlight)
       )
plt.title(item_to_highlight)
plt.ylabel("number of fruit")

plt.show()
````

````python
# 3. same as above, but with a function. Notice tha I made the dictionary a variable to, so that is is even more universal
def make_a_plot(dict_to_visualise, which_item, item_to_highlight, colour_to_highlight):
        plt.subplot(3,1,which_item)
        plt.bar(dict_to_visualise.keys(),
                dict_to_visualise.values(),
                color = colour_for_highlight(dict_to_visualise.keys(), item_to_highlight, colour_to_highlight)
        )
        plt.title(item_to_highlight)
        plt.ylabel("number of fruit")

make_a_plot(sorted_fruit_counts, 1,"banana", "yellow")
plt.show()
````

````python
# 3. and finally all 3:

# same as above, but with a function. Notice tha I made the dictionary a variable to, so that is is even more universal
make_a_plot(sorted_fruit_counts, 1,"banana", "yellow")
make_a_plot(sorted_fruit_counts, 2,"plum", "purple")
make_a_plot(sorted_fruit_counts, 3,"quince", "orange")
plt.show()
````
